# Contrast Stretching:

The Lower value of the input image is mapped to 0 and the upper value is mapped to 255. All other intermediate values are reassigned new intensity values according to the following formula



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
pic1 = cv2.imread("forest.jpg")
```

```python
plt.imshow(pic1)
```

```python
plt.hist(pic1.ravel(),256,[0,300]); plt.show()
```

```python
pic1
```

```python
xmax = np.max(pic1)
```

```python
xmax
```

```python
xmin = np.min(pic1)
```

```python
xmin
```

```python
minmax_image = (pic1-xmin)/(xmax-xmin)
```

```python
minmax_image
```

```python
plt.imshow(minmax_image)
```

```python
plt.hist(minmax_image.ravel(),256,[0,300]); plt.show()
```

# Thersolding:

Above the thersold value it will be converted to 255 below threshold to 0

```python
img = cv2.imread('sunflower.jpg') 
```

```python
plt.imshow(img) 
plt.show() 
```

```python
hist = cv2.calcHist([img],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show()
```

```python
ret,thresh1 = cv2.threshold(img,128,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh1) 
plt.show() 
```

```python
hist = cv2.calcHist([thresh1],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show()
```
