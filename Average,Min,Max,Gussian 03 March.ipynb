# Averaging Filter:

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Read the image 
img = cv2.imread('parrot.jpg',0) 
plt.imshow(img)
```

```python
# Obtain number of rows and columns of the image 
m, n = img.shape 
   
# Develop Averaging filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9
   
```

```python
# Convolve the 3X3 mask over the image  
img_new = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
         
        img_new[i, j]= temp 
```

```python
img_new = img_new.astype(np.uint8) 
plt.imshow(img_new)
```

# Median filter:

It is also known as nonlinear filtering.The pixel value is replaced by the median value of the neighboring pixel.

```python
# Read the image 
img_noise = cv2.imread('parrot.jpg', 0)
plt.imshow(img_noise)
# Obtain the number of rows and columns of the image 
m, n = img_noise.shape 
```

```python
# Traverse the image. For every 3X3 area,find the median of the pixels and 
# replace the ceter pixel by the median 
img_new1 = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = [img_noise[i-1, j-1], 
               img_noise[i-1, j], 
               img_noise[i-1, j + 1], 
               img_noise[i, j-1], 
               img_noise[i, j], 
               img_noise[i, j + 1], 
               img_noise[i + 1, j-1], 
               img_noise[i + 1, j], 
               img_noise[i + 1, j + 1]] 
          
        temp = sorted(temp) 
        img_new1[i, j]= temp[4] 
```

```python
img_new1 = img_new1.astype(np.uint8) 
plt.imshow(img_new1)
```

# Min filter:

```python
# Read the image 
img = cv2.imread('parrot.jpg', 0)
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1,j-1]*mask[0,0],img[i-1,j]*mask[0,1],img[i-1,j+1]*mask[0,2],img[i,j-1]*mask[1,0],img[i,j]*mask[1,1],img[i,j+1]*mask[1,2],img[i+1,j-1]*mask[2,0],img[i+1,j]*mask[2,1],img[i+1,j+1]*mask[2,2]
        a = np.sort(temp)
        minimum = np.min(a)
        img_new[i,j] = minimum
```

```python
plt.imshow(img_new)
```

# Max filter:

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1,j-1]*mask[0,0],img[i-1,j]*mask[0,1],img[i-1,j+1]*mask[0,2],img[i,j-1]*mask[1,0],img[i,j]*mask[1,1],img[i,j+1]*mask[1,2],img[i+1,j-1]*mask[2,0],img[i+1,j]*mask[2,1],img[i+1,j+1]*mask[2,2]
        a = np.sort(temp)
        maximum = np.max(a)
        img_new[i,j] = maximum
```

```python
plt.imshow(img_new)
```
