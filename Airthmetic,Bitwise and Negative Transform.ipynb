# Negative Image using PIL

```python
import numpy as np
from PIL import Image
from PIL import ImageFilter
```

```python
img=Image.open('imag1.jpg')
```

```python
img.size
img.size[0]
img.size[1]
```

```python
#0-255 0 indicate least(black) 255 indicate highest (white)
#negative transformation so that color gets inverted
for i in range(0,img.size[0]-1):
    for j in range(0,img.size[1]-1):
        pixelColorVals=img.getpixel((i,j));
        
        redPixel=255-pixelColorVals[0];
        greenPixel=255-pixelColorVals[1];
        bluePixel=255-pixelColorVals[2];
        
        img.putpixel((i,j),(redPixel,greenPixel,bluePixel))
        
```

# Negative Image using CV2

```python
img=cv2.imread('imag1.jpg')
```

```python
plt.imshow(img)
```

```python
height,width,_ =img.shape 
```

```python
for i in range(0, height - 1): 
    for j in range(0, width - 1): 
        pixel = img[i, j] 
        
        pixel[0] = 255 - pixel[0] 
        
        pixel[1] = 255 - pixel[1] 
        
        pixel[2] = 255 - pixel[2] 
        img[i, j] = pixel
```

```python
plt.imshow(img)
```
